```python
import pandas as pd
```

```python
df = pd.read_excel('projeto1A.xlsx')
```

```python
df = df.fillna(0)
```

```python
df.set_index('Nome do usuário', inplace=True)
```

```python
df.loc[df['README'] == 0, 'README'] = 'Repositório não possui README.'
```

```python
df.loc[df['README'] == 'x', 'README'] = 'Repositório possui arquivo README.'
```

```python
df.loc[df['Estilo Página'] == 'x', 'Estilo Página'] = 'Possui estilo CSS.'
```

```python
df.loc[df['Editar Anotações'] == 'x', 'Editar Anotações'] = 'Implementou edição de notas.'
df.loc[df['Editar Anotações'] == 0, 'Editar Anotações'] = 'Não implementou edição de notas.'
```

```python
df.loc[df['Remover Anotacão'] == 'x', 'Remover Anotacão'] = 'Implementou opção de deletar.'
df.loc[df['Remover Anotacão'] == 0, 'Remover Anotacão'] = 'Não implementou opção de deletar.'
```

```python
df.loc[df['SQLite'] == 'x', 'SQLite'] = 'Utilizou SQlite'
```

```python
df.loc[df['Mesma Página'] == 'x', 'Mesma Página'] = 'Implementou todas as funcionalidades na mesma página.'
df.loc[df['Mesma Página'] == 0, 'Mesma Página'] = 'Não implementou todas as funcionalidades na mesma página.'
```

```python
df.loc[df['Adicional'] == 'x', 'Adicional'] = 'Implementou alguma funcionalidade adicional.'
df.loc[df['Adicional'] == 0, 'Adicional'] = 'Não implementou funcionalidade adicional.'
```

```python
colunas = list(df.columns)
colunas
```

```python
feedback = colunas.copy()
feedback.remove('Nome')
feedback.remove('Sobrenome')
feedback.remove('Link GitHub')
feedback.remove('Conceito')
feedback.remove('OBS')
```

```python
for index, row in df.iterrows():
    feedback_texto = ''
    
    for coluna in feedback:
        feedback_texto += str(row.loc[coluna]) + '\n'
    
    feedback_texto += f'Obs. Adicionais: {row.loc["OBS"]} \n'
    feedback_texto += f'Conceito: {row.loc["Conceito"]}'
    
    df.loc[index, 'feedback_completo'] = feedback_texto
    print(feedback_texto)
    print()
    print('*'*50)
    
```
