```python
import pandas as pd
```

```python
df = pd.read_excel('2023-1-sprint1-responses.xlsx')
df.set_index('Selecione o seu Nome', inplace=True)
df.drop(['Timestamp'], axis=1, inplace=True)
df.drop(['Email Address'], axis=1, inplace=True)
```

```python
columns = df.columns
```

```python
new_columns = {}
for column in columns:
    # posição das palavras 'PRODUTIVIDADE', 'PROATIVIDADE', 'TRANSPARÊNCIA' no nome da coluna
    categoria = column.split()[3]
    nome = column.split('[')[-1].replace(']','')
    #removendo texto desnecessário da coluna e renomeando no padrão ex: PRODUTIVIDADE - NOME DO ALUN@
    new_columns[column] = f'{categoria} - {nome}'

```

```python
df.rename(columns=new_columns, inplace=True)
```

```python
responder = []
```

```python
transposed = df.T
for aluno in df.index:

    index_filter = [index for index in transposed.index if 'PRODUTIVIDADE' in index]
    if 5 in list(transposed.loc[index_filter, aluno].dropna()):
        responder.append(aluno)
```
