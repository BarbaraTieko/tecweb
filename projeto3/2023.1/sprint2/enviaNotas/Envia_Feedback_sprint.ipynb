```python
password = "eeikpwsjhetneesg"


def envia(alunos):
    alunos.set_index(usercol, inplace=True)
    
    # me == my email address
    # you == recipient's email address

    dados = alunos


    for id_aluno, seq in dados.iterrows():

        text = seq['feedback']

        email = id_aluno

        me = "barbarata@insper.edu.br"

        if debugMode:
            you = "barbara.agena@al.insper.edu.br"
        else:
            you = email + "@al.insper.edu.br" #### NEutralize
            
        print(f'Enviando email -> {you}')

        # Create message container - the correct MIME type is multipart/alternative.
        msg = MIMEMultipart('alternative')
        msg['Subject'] = "TECWEB: Avaliação Individual dos Membros - Sprint 1"
        msg['From'] = me
        msg['To'] = you    

        part1 = MIMEText(text, 'plain')
    
        # Attach parts into message container.
        # According to RFC 2046, the last part of a multipart message, in this case
        # the HTML message, is best and preferred.
        msg.attach(part1)

        # Send the message via local SMTP server.
        #mail = smtplib.SMTP('smtp.gmail.com', 587)
        mail = smtplib.SMTP_SSL('smtp.gmail.com', 465)

        mail.login('tecnologias.web.insper', password)
        mail.sendmail(me, you, msg.as_string())
        mail.quit()
    
    print('Todos os emails enviados!!!')
```
