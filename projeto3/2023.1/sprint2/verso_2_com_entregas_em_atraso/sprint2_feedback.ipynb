```python
import pandas as pd
```

```python
df = pd.read_excel('sprint2_medias.xlsx')
```

```python
df.set_index('nome', inplace=True)
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
def texto_diferenca(diferenca):
    if diferenca == 0:
        return 'Ou seja, a visão dos colegas sobre você foi igual.\n    Sua percepção parece correta.'
    elif diferenca > 0:
        return 'Ou seja, a visão dos colegas sobre você foi superior.'
    elif diferenca < 0 and diferenca <= -1:
        return 'Ou seja, a visão dos colegas sobre você foi um pouco inferior.\n    Não muito, mas vale uma reflexão.'
    else:
        return 'Ou seja, a visão dos colegas sobre você foi um pouco inferior, vale uma reflexão.'
```

```python
def generate_feedback(aluno):
    first_name = aluno.split()[0].capitalize()
    AUTO_PRODUTIVIDADE = round(df.loc[aluno,"auto_produtividade"], 2)
    MEDIA_PRODUTIVIDADE = round(df.loc[aluno,"media_produtividade"],2)
    DIFERENCA_PRODUTIVIDADE = round(MEDIA_PRODUTIVIDADE - AUTO_PRODUTIVIDADE ,2)
    
    AUTO_PROATIVIDADE = round(df.loc[aluno,"auto_proatividade"], 2)
    MEDIA_PROATIVIDADE = round(df.loc[aluno,"media_proatividade"],2)
    DIFERENCA_PROATIVIDADE = round(MEDIA_PROATIVIDADE - AUTO_PROATIVIDADE,2)
    
    AUTO_TRANSPARENCIA = round(df.loc[aluno,"auto_transparência"], 2)
    MEDIA_TRANSPARENCIA = round(df.loc[aluno,"media_transparência"],2)
    DIFERENCA_TRANSPARENCIA = round(MEDIA_TRANSPARENCIA - AUTO_TRANSPARENCIA,2)
    
    texto = f'''
    Olá {first_name}, tudo bem?

    Segue abaixo seu feedback de grupo da Sprint 2.

    Lembrete: Um membro é considerado bem avaliado quando obtém uma média entre 3 e 5 em produtividade,
    uma média entre 3 e 4 em proatividade em uma média entre 3 e 4 em transparência.

    A diferença entre 3 e 4, sozinha, não faz diferença para
    nota. O nível 4 deve ser usado para reconhecer o esforço
    de membros específicos.

    Produtividade:
    -------------

    Nível que você atribuiu a si mesmo: {AUTO_PRODUTIVIDADE}

    Média dos níveis que os colegas atribuíram a você: {MEDIA_PRODUTIVIDADE}

    Diferença: {DIFERENCA_PRODUTIVIDADE}

    {texto_diferenca(DIFERENCA_PRODUTIVIDADE)}


    Proatividade:
    ------------

    Nível que você atribuiu a si mesmo: {AUTO_PROATIVIDADE}

    Média dos níveis que os colegas atribuíram a você: {MEDIA_PROATIVIDADE}

    Diferença: {DIFERENCA_PROATIVIDADE}

    {texto_diferenca(DIFERENCA_PROATIVIDADE)}


    Transparência
    -------------

    Nível que você atribuiu a si mesmo: {AUTO_TRANSPARENCIA}

    Média dos níveis que os colegas atribuíram a você: {MEDIA_TRANSPARENCIA}

    Diferença: {DIFERENCA_TRANSPARENCIA}

    {texto_diferenca(DIFERENCA_TRANSPARENCIA)}
    
    
    Esteja à vontade para perguntar em caso de quaisquer dúvidas.
    
    Abraços,
    Bárbara Agena

'''
    return texto
    
```

```python
data = {
    'nome':[],
    'feedback':[]
    
}
```

```python
alunos = list(df.index)
for aluno in alunos:
    data['nome'].append(aluno)
    data['feedback'].append(generate_feedback(aluno))
```
